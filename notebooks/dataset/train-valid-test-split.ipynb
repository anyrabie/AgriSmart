```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import shutil
import random
```

```python
print("GPU Available: ",tf.config.list_physical_devices('GPU'))
```

```python
def organize_dataset(source, dest, train_r=0.7, valid_r=0.15):

    for split in ['train', 'valid', 'test']:
        os.makedirs(f'{dest}/{split}', exist_ok=True)
    for class_name in os.listdir(source):
        class_path = os.path.join(source, class_name)

        if not os.path.isdir(class_path):
            continue

        for split in ['train', 'valid', 'test']:
            os.makedirs(f'{dest}/{split}/{class_name}', exist_ok=True)

        images = [f for f in os.listdir(class_path)
                 if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
        random.seed(42)
        random.shuffle(images)
        n = len(images)
        n_train = int(n * train_r)
        n_valid = int(n * valid_r)

        train_imgs = images[:n_train]
        valid_imgs = images[n_train:n_train+n_valid]
        test_imgs = images[n_train+n_valid:]

        for img in train_imgs:
            shutil.copy(f'{class_path}/{img}', f'{dest}/train/{class_name}/{img}')

        for img in valid_imgs:
            shutil.copy(f'{class_path}/{img}', f'{dest}/valid/{class_name}/{img}')

        for img in test_imgs:
            shutil.copy(f'{class_path}/{img}', f'{dest}/test/{class_name}/{img}')

        print(f"{class_name}: Train={len(train_imgs)},"f"Valid={len(valid_imgs)}, Test={len(test_imgs)}")
```

```python
source_path = '/kaggle/input/datasets/abdallahalidev/plantvillage-dataset/segmented'
output_path = '/kaggle/working/plantvillage_split'
organize_dataset(source_path, output_path)
print("\n RÃ‰SUMÃ‰:")
print(f"Train: {len(os.listdir(f'{output_path}/train'))} classes")
print(f"Valid: {len(os.listdir(f'{output_path}/valid'))} classes")
print(f"Test: {len(os.listdir(f'{output_path}/test'))} classes")

# Compter le nombre total d'images
total_train = sum([len(os.listdir(f'{output_path}/train/{c}')) 
                   for c in os.listdir(f'{output_path}/train')])
total_valid = sum([len(os.listdir(f'{output_path}/valid/{c}')) 
                   for c in os.listdir(f'{output_path}/valid')])
total_test = sum([len(os.listdir(f'{output_path}/test/{c}')) 
                  for c in os.listdir(f'{output_path}/test')])

print(f"\nðŸ“¸ IMAGES:")
print(f"Train: {total_train} images")
print(f"Valid: {total_valid} images")
print(f"Test: {total_test} images")
print(f"TOTAL: {total_train + total_valid + total_test} images")

print("\n Preprocessing terminÃ© ! PrÃªt pour la crÃ©ation du dataset.")
```
